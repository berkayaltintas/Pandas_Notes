# Gruplaştırma İşlemleri (10) # 

```python
import pandas as pd
```

```python
df = pd.DataFrame({'gruplar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'veri': [10,11,52,23,43,55]}, columns=['gruplar', 'veri'])
df
```

```python
df.groupby("gruplar")
```

```python
df.groupby("gruplar").mean()
```

A B C değerlerinin ortalaması alınmıştır.

```python
df.groupby("gruplar").sum()
```

A B C değerleri toplanmıştır.

```python
import seaborn as sns
df = sns.load_dataset("planets")
df.head() #Baştaki verileri getirdik.
```

```python
df.groupby("method")
```

```python
df.groupby("method")["orbital_period"].mean()
```

```python
df.groupby("method")["orbital_period"].describe()
```

orbital_period burada bir değişkendir.Değişken ile beraber kullanılmalıdır.

```python
df.groupby("method")["orbital_period"].describe()
```
