# Birleştirme İşlemleri (Join İşlemleri) (7) #

```python
import numpy as np
import pandas as pd
a=np.random.randint(1,30 , size=(5,3))
df1=pd.DataFrame(a,columns=["sutun1","sutun2","sutun3"])
```

```python
df1
```

```python
df2=df1+91 
```

```python
df2
```

Şimdi bu iki tabloyu birleştirme işlemi yapalım.

```python
pd.concat([df1,df2]) #concat fonksiyonu birleştirme işlemi yapılır.
```

Bu iki tabloyu birleştirdik ve görüldüğü üzere indexler düzgün değil.Bunu düzeltelim.

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
#  ignore_index=True yaparak indexleri düzeltiyoruz.
```

```python
df1.columns
```

**Sütun isimlerini de değiştirebiliriz.**

```python
df2.columns=["sutun1","sutun2","sutun4"]
```

```python
df2
```

-->Şimdi df1 ve df2 ile tekrar birleştirelim. (Artık sütun isimlerinden biri farklı)

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True) #-->İndexleri düzeltelim.
```

```python
pd.concat([df1,df2],join="inner")
```

_İnner işlemi (KESİŞİM) olarak alındı._

```python
pd.concat([df1,df2],join="inner", ignore_index=True)
```
